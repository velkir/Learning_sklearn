```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder


data = {
    "products": [
        "bread eggs",
        "bread eggs milk",
        "milk cheese",
        "bread butter cheese",
        "eggs milk",
        "bread milk butter cheese",
    ]
}

transactions = pd.DataFrame(data)
customers_purchases = transactions["products"].str.split(pat=" ", expand=True, n=4)

columns = sorted(customers_purchases.stack().unique())
rows_len = len(customers_purchases.index)
columns_len = len(columns)
dummies = pd.DataFrame(data=np.zeros(shape=(rows_len, columns_len)), dtype=int, columns=columns)

for product in columns:
  for idx, row_data in customers_purchases.iterrows():
    if any(row_data.isin([product])):
      dummies[product][idx]=1

support_butter_bread = round(len(dummies[(dummies["bread"] == 1) & (dummies["butter"] == 1)])/len(dummies), 4)
support_butter_milk = round(len(dummies[(dummies["milk"] == 1) & (dummies["butter"] == 1)])/len(dummies), 4)

print("Support butter and bread: {}".format(support_butter_bread))
print("Support butter and milk: {}".format(support_butter_milk))
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder


data = {
    "products": [
        "bread eggs",
        "bread eggs milk",
        "milk cheese",
        "bread butter cheese",
        "eggs milk",
        "bread milk butter cheese",
    ]
}

transactions = pd.DataFrame(data)
customers_purchases = transactions["products"].str.split(pat=" ", expand=True, n=4)

columns = sorted(customers_purchases.stack().unique())
rows_len = len(customers_purchases.index)
columns_len = len(columns)
dummies = pd.DataFrame(data=np.zeros(shape=(rows_len, columns_len)), dtype=int, columns=columns)

for product in columns:
  for idx, row_data in customers_purchases.iterrows():
    if any(row_data.isin([product])):
      dummies[product][idx]=1

conf_cheese_bread = len(dummies[(dummies["cheese"] == 1) & (dummies["bread"] == 1)])/len(dummies[dummies["cheese"] == 1])
conf_butter_cheese = len(dummies[(dummies["butter"] == 1) & (dummies["cheese"] == 1)])/len(dummies[dummies["butter"] == 1])

print(f"conf(cheese, bread) = {conf_cheese_bread:.4f}")
print(f"conf(butter, cheese) = {conf_butter_cheese:.4f}")
```
